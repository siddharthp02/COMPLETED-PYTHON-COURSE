```python
import re
```

```python
for folder, sub_folders , files in os.walk('extracted__content'):
    print(f'currently in folder: {folder}')
    print('\n')
    print(f'the subfolders are:')
    for sub_folder in sub_folders:
        print('\t'+sub_folder)
    print('\n')
    print('the files are:')
    for file in files:
        print('\t'+file)
        
        my_file = open(f'{os.getcwd()}/{folder}/{file}')
       
        print('\n\t\t\tMatches: ')
        
        
        for match in re.finditer(r'\d{3}-\d{3}-\d{4}',my_file.read()):
            print(f'\t\t\t\t{match.group()}')
            
    print('\n')
    
    
```

```python
for folder, sub_folders , files in os.walk('extracted__content'):
    for file in files:
        
        my_file = open(f'{os.getcwd()}/{folder}/{file}')
        
        
        for match in re.finditer(r'\d{3}-\d{3}-\d{4}',my_file.read()):
            print(f'Match:{match.group()}')
            print(f'{os.getcwd()}/{folder}/{file}')
            
```
